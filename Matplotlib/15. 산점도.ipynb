## 15. 산점도 그래프 

```python
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자기 깨지는 현상 해결
```

```python
# 엑셀 파일 데이터 불러오기
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
import numpy as np

sizes = np.random.rand(8) * 1000
sizes
```

```python
# dot size 변경
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
# 학년에 따라 dot size 변경
sizes = df['학년'] * 200
sizes

plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
# dot color 변경
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha= 0.5)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

👀 색 참고
> [https://matplotlib.org/stable/tutorials/colors/colormaps.html](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
plt.figure(figsize=(7, 7))

plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha= 0.5)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')

plt.colorbar(ticks=[1,2,3], label='학년', shrink = 0.7, orientation='horizontal') # 컬러 바 추가
```
