## 8. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자기 깨지는 현상 해결
```

```python
labels = ['짱구', '철수', '유리']
values = [80, 84, 77]

plt.bar(labels, values)
```

```python
# 막대 색 변경
plt.bar(labels, values, color = 'g')
```

```python
# 데이터별 막대 색깔
colors = ['r', 'b', 'pink'] 

plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
# y축 범위 설정
plt.bar(labels, values)
plt.ylim(60, 90) 
```

```python
# 막대 두께 설정
plt.bar(labels, values, width=0.5)
```

```python
# x, y축의 데이터 각도를 45도로 설정
plt.bar(labels, values, width=0.3)
plt.xticks(rotation= 45)
plt.yticks(rotation= 45)
```

```python
labels = ['짱구', '철수', '유리']
values = [80, 84, 77]
ticks = ['1번학생', '2번학생', '3번학생']

plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=45)
```
