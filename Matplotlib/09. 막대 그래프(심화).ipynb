```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # Í∏ÄÏûêÌÅ¨Í∏∞
matplotlib.rcParams['axes.unicode_minus'] = False # ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÇ¨Ïö© Ïãú, ÎßàÏù¥ÎÑàÏä§ Í∏ÄÏûêÍ∏∞ Íπ®ÏßÄÎäî ÌòÑÏÉÅ Ìï¥Í≤∞
```

```python
# ÏàòÌèâ ÎßâÎåÄÍ∑∏ÎûòÌîÑ
labels = ['Ïß±Íµ¨', 'Ï≤†Ïàò', 'Ïú†Î¶¨']
values = [80, 84, 77]

plt.barh(labels, values)
plt.xlim(60, 90) # xÏ∂ï Î≤îÏúÑ ÏÑ§Ï†ï
```

### ‚óΩ hatch style

üëÄ Ï∞∏Í≥†
> [https://matplotlib.org/3.5.0/gallery/shapes_and_collections/hatch_style_reference.html](https://matplotlib.org/3.5.0/gallery/shapes_and_collections/hatch_style_reference.html)

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/')
bar[1].set_hatch('**')
bar[2].set_hatch('..')
```

```python
bar = plt.bar(labels, values)
plt.ylim(60, 90)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height()+0.5, values[idx], ha='center', color='b')
```
