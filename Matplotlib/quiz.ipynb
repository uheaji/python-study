```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

data = {
    'ÏòÅÌôî' : ['Î™ÖÎüâ', 'Í∑πÌïúÏßÅÏóÖ', 'Ïã†Í≥ºÌï®Íªò-Ï£ÑÏôÄ Î≤å', 'Íµ≠Ï†úÏãúÏû•', 'Í¥¥Î¨º', 'ÎèÑÎëëÎì§', '7Î≤àÎ∞©Ïùò ÏÑ†Î¨º', 'ÏïîÏÇ¥'],
    'Í∞úÎ¥â Ïó∞ÎèÑ' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    'Í¥ÄÍ∞ù Ïàò' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (Îã®ÏúÑ : Îßå Î™Ö)
    'ÌèâÏ†ê' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

#### üö® ÏòÅÌôî Îç∞Ïù¥ÌÑ∞Î•º ÌôúÏö©ÌïòÏó¨ xÏ∂ïÏùÄ ÏòÅÌôî, yÏ∂ïÏùÄ ÌèâÏ†êÏù∏ ÎßâÎåÄ Í∑∏ÎûòÌîÑÎ•º ÎßåÎìúÏãúÏò§. 

```python
plt.bar(df['ÏòÅÌôî'], df['ÌèâÏ†ê'])
```

#### üö® ÏúÑ Í∑∏ÎûòÌîÑÏóê Ï†úÏãúÎêú ÏÑ∏Î∂ÄÏÇ¨Ìï≠ÏùÑ Ï†ÅÏö©ÌïòÏãúÏò§.

- Ï†úÎ™© : Íµ≠ÎÇ¥ Top 8 ÏòÅÌôî ÌèâÏ†ê Ï†ïÎ≥¥
- XÏ∂ï label : ÏòÅÌôî (90ÎèÑ ÌöåÏ†Ñ)
- YÏ∂ï label : ÌèâÏ†ê 

```python
plt.bar(df['ÏòÅÌôî'], df['ÌèâÏ†ê'])
plt.title('Íµ≠ÎÇ¥ Top 8 ÏòÅÌôî ÌèâÏ†ê Ï†ïÎ≥¥')
plt.xlabel('ÏòÅÌôî')
plt.xticks(rotation=90)
plt.ylabel('ÌèâÏ†ê')
```

#### üö® Í∞úÎ¥â Ïó∞ÎèÑÎ≥Ñ ÌèâÏ†ê Î≥ÄÌôîÏ∂îÏù¥Î•º Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑÎ°ú Í∑∏Î¶¨ÏãúÏò§.

- Ïó∞ÎèÑÎ≥Ñ ÌèâÍ∑† Îç∞Ïù¥ÌÑ∞Î•º Íµ¨ÌïòÎäî ÏΩîÎìú

```python
df_group = df.groupby('Í∞úÎ¥â Ïó∞ÎèÑ').mean()
df_group 
```

```python
plt.plot(df_group.index, df_group['ÌèâÏ†ê'])
```

#### üö® ÏúÑ Í∑∏ÎûòÌîÑÏóê Ï†úÏãúÎêú ÏÑ∏Î∂ÄÏÇ¨Ìï≠ÏùÑ Ï†ÅÏö©ÌïòÏãúÏò§.

- marker : 'o'
- xÏ∂ï ÎààÍ∏à : 5ÎÖÑ Îã®ÏúÑ (2005, 2010, 2015, 2020)
- yÏ∂ï Î≤îÏúÑ : ÏµúÏÜå 7, ÏµúÎåÄ 10 

```python
plt.plot(df_group.index, df_group['ÌèâÏ†ê'], marker='o')
plt.xticks([2005,2010,2015,2020])
plt.ylim(7, 10)
```

#### üö® ÌèâÏ†êÏù¥ 9Ï†ê Ïù¥ÏÉÅÏù∏ ÏòÅÌôîÏùò ÎπÑÏú®ÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÎäî Ïõê Í∑∏ÎûòÌîÑÎ•º Ï†úÏãúÎêú ÏÑ∏Î∂ÄÏÇ¨Ìï≠ÏùÑ Ï†ÅÏö©ÌïòÏó¨ Í∑∏Î¶¨ÏãúÏò§. 
(Î≥µÏäµÌïòÍ∏∞...üò•)

- label : 9Ï†êÏù¥ÏÉÅ / 9Ï†êÎØ∏Îßå
- ÌçºÏÑºÌä∏ : ÏÜåÏàòÏ†ê Ï≤´Ïß∏ÏûêÎ¶¨ÍπåÏßÄ ÌëúÏãú
- Î≤îÎ°Ä : Í∑∏ÎûòÌîÑ Ïö∞Ï∏°Ïóê ÌëúÏãú 

```python
filt = df['ÌèâÏ†ê'] >= 9.0
values = [len(df[filt]), len(df[-filt])]
labels = ['9Ï†ê Ïù¥ÏÉÅ', '9Ï†ê ÎØ∏Îßå']

plt.pie(values, labels=labels, autopct='%.1f%%')
plt.legend(loc=(1,0.3))
plt.show()
```
