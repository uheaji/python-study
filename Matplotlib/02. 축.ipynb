## 2. 축 

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자기 깨지는 현상 해결
```

```python
x =[1, 2, 3]
y =[2, 6, 4]
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold', 'size':20}) # 개별폰트 설정
```

```python
# 레이블 추가
plt.plot(x,y)
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
# 레이블에 색추가
plt.plot(x,y)
plt.xlabel('X축', color='red')
plt.ylabel('Y축', color='#00aa00')
```

```python
# 레이블 위치
plt.plot(x,y)
plt.xlabel('X축', color='red', loc='right') # left, center, right
plt.ylabel('Y축', color='#00aa00',  loc='top') # top, center, bottom
```

```python
# 축 tick 정의
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3, 6, 9, 12])
plt.show()
```
