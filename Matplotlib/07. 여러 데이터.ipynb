```python
days = [1, 2, 3] # 1Ïùº, 2Ïùº, 3Ïùº
az = [2, 4, 8] # (Îã®ÏúÑ: Îßå Î™Ö) 1ÏùºÎ∂ÄÌÑ∞ 3ÏùºÍπåÏßÄ ÏïÑÏä§Ìä∏ÎùºÏ†úÎÑ§Ïπ¥ Ï†ëÏ¢ÖÏù∏Íµ¨
pfizer = [5, 1, 3] # ÌôîÏù¥Ïûê
moderna = [1, 2, 6] # Î™®ÎçîÎÇò

plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
```
