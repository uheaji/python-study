## 3. 범례 

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자기 깨지는 현상 해결
```

```python
x =[1, 2, 3]
y =[2, 6, 4]
```

```python
plt.plot(x,y, label='무슨 데이터')
plt.legend()
```

```python
# 범례위치 변경
plt.plot(x,y, label='무슨 데이터')
plt.legend(loc='upper right') # 오른쪽 위
```

```python
plt.plot(x,y, label='무슨 데이터')
plt.legend(loc='lower right') # 오른쪽 아래
```

```python
plt.plot(x,y, label='범례')
plt.legend(loc=(0.5, 0.5)) # x축, y축 (0~1사이)
```
