### 10. 결측치
비어 있는 데이터 

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

#### 데이터 채우기 fillna 

```python
df.fillna('') # NaM 데이터를 빈 칸으로 채움
```

```python
df.fillna('없음') 
```

```python
import numpy as np
df['소속'] = np.nan # 소속 데이터 전체를 NaN으로 채움\
df
```

```python
df.fillna('모름', inplace=True)
```

```python
df # '모름'으로 처리가 안되어있음 ! => inplace
```

```python
df['SW특기'].fillna('확인 중', inplace=True) # SW특기 데이터 중에서 NaN에 대해서 채움
df
```

#### 데이터 제외하기 dropna 

```python
df.dropna(inplace= True) # 전체 데이터 중에서 NaN를 포함하는 데이터 삭제
df
```

- axis : index or columns 
- how : any or all

```python
df.dropna(axis='index', how='any') # NaN이 하나라도 있는 row 삭제
```

```python
df.dropna(axis='columns') # NaN이 하나라도 있는 column 삭제 (SW특기 컬럼이 사라짐)
```

```python
df['소속'] = np.nan
df
```

```python
df.dropna(axis='columns', how='all') # 컬럼데이터 전체가 NaN이면 column 삭제 
```
