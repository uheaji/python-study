### 14. 그룹화 
동일한 값을 가진 것들끼리 합쳐서 통계 또는 평균 등의 값을 계산하기 위해 사용 

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.groupby('소속') 
```

```python
df.groupby('소속').get_group('떡잎유치원')
```

```python
df.groupby('소속').mean() # 계산 가능한 데이터들의 평균값 
```

```python
df.groupby('소속').size() # 각 그룹의 크기
```

```python
df.groupby('소속').size()['없음'] # 소속으로 그룹화를 한 뒤에 '없음'에 해당하는 데이터의 수
```

```python
df.groupby('소속')['키'].mean()  # 소속으로 그룹화를 한 뒤에 키의 평균 데이터 
```

```python
df.groupby('소속')[['국어', '영어', '수학']].mean() # 소속으로 그룹화를 한 뒤에 국어, 영어, 수학의 평균 데이터
```

```python
df['나이'] = [5, 34, 29, 1, 5, 5, 5, 5] # 나이 column 추가
df
```

```python
df.groupby(['소속', '나이']).mean() # 소속별, 학년별 평균 데이터
```

```python
df.groupby('나이').mean()
```

```python
df.groupby('나이').mean().sort_values('키') # 나이별로 그룹화하여 평균데이터를 나타내고 키를 기준으로 오름차순으로 정렬
```

```python
df.groupby(['소속', '나이']).sum()
```

```python
df.groupby(['나이'])[['이름', 'SW특기']].count() # 나이별로 그룹화 한 뒤 SW특기의 개수를 출력
```

```python
school = df.groupby('소속')
school['나이'].value_counts() # 소속으로 그룹화 한 뒤에 나이별로 학생 수를 가져옴
```
