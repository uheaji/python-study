## Pandas
파이썬에서 사용하는 데이터 분석 라이브러리

```python
import pandas as pd
```

### 1.Series
1차원 데이터 (정수, 실수, 문자열 등)

#### Series 객체생성
예) 1월부터 4월까지 평균 온도데이터(-20, -10, 10, 20)

```python
temp = pd.Series([-20, -10, 10, 20])
print(temp)
```

```python
temp[0] # 1월온도
temp[2] # 3월온도
```

#### Series 객체생성 (index 지정)

```python
temp = pd.Series([-20, -10, 10, 20], index=['Jan', 'Feb', 'Mar', 'Apr'])
temp
```

```python
temp['Jan'] # index Jan(1월)에 해당하는 데이터 출력
```

```python
temp['Apr'] # index Apr(4월)에 해당하는 데이터 출력
```
