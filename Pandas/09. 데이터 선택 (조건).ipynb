### 9. 데이터 선택 (조건) 
조건에 해당하는 데이터 선택 

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 180 # 학생들의 키가 180 이상인지 여부(True/False)
```

```python
filt = (df['키'] >= 180)
df[filt] # 조건으로 데이터프레임 조회
```

```python
df[-filt] # 조건을 역으로 적용하여 조회
```

```python
df[df['키'] >= 180]
```

```python
df.loc[df['키'] >= 180, '수학']
```

```python
df.loc[df['키'] >= 180, ['수학', '과학']]
```

#### 다양한 조건

##### & 그리고

```python
df.loc[(df['키'] <= 180) & (df['소속'] == '떡잎유치원')]
```

##### | 또는 

```python
df.loc[(df['키'] <= 180) | (df['소속'] == '떡잎상사')]
```

#### str 함수 

```python
filt = df['이름'].str.startswith('신') # '신'씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('구') # 이름에 '구'가 들어가는 사람
df[filt]
```

```python
df[-filt] # 이름에 '구'가 들어가는 사람 제외
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs)
df[filt] # SW특기가 파이썬이거나 자바인 사람
```

```python
df # 유리도 파이썬인데 모두 대문자라서 출력안됨
```

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs) # 소문자로 변환
df[filt] # SW특기가 파이썬이거나 자바인 사람
```

```python
df['SW특기'].str.contains('Java', na =True) # NaN 데이터에 대해서 True로 설정
```

```python
df['SW특기'].str.contains('Java', na =False) # NaN 데이터에 대해서 False로 설정
```

```python
filt = df['SW특기'].str.contains('Java', na=False)
df[filt]
```
