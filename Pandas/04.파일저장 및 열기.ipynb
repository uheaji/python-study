## 4. 파일 저장 및 열기
DataFrame 객체를 Excel, csv, txt등의 형태의 파일로 저장 및 열기 

```python
import pandas as pd
data = {
    '이름' : ['신짱구', '신형만', '봉미선', '신짱아', '훈이', '철수', '유리', '맹구'],
    '소속' : ['떡잎유치원', '떡잎상사', '없음', '없음', '떡잎유치원', '떡잎유치원', '떡잎유치원', '떡잎유치원'],
     '키' : [80, 184, 168, 60, 75, 80, 74, 85],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}

df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번'])
df
```

```python
df.index.name = '지원번호'
df
```

***
### 저장하기
#### csv파일로 저장  

```python
df.to_csv('score.csv', encoding='utf-8-sig')
```

```python
# 인덱스는 제외하고 파일저장
df.to_csv('score.csv', encoding='utf-8-sig', index=False)
```

#### 텍스트(.txt)파일로 저장 

```python
df.to_csv('score.txt', sep='\t') # tab으로 구분된 텍스트 파일
```

#### 엑셀 파일로 저장 

```python
df.to_excel('score.xlsx')
```

***
### 파일열기 

#### csv 파일 열기 

```python
df = pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows=1) # 지정된 갯수 만큼의 행을 건너뜀
df
```

```python
df = pd.read_csv('score.csv', skiprows=[1, 3, 5]) # 1, 3, 5행 제외 (row는 0부터 시작)
df
```

```python
df = pd.read_csv('score.csv', nrows=4) # 지정된 개수만큼의 행만 가져옴
df
```

```python
df = pd.read_csv('score.csv', skiprows=2, nrows=4) # 처음 2개행은 무시, 이후에 4개의 행을 가져옴
df
```

#### 텍스트 파일 열기 

```python
df = pd.read_csv('score.txt', sep='\t')
df
```

```python
df = pd.read_csv('score.txt', sep='\t', index_col='지원번호') # 인덱스를 지원번호 컬럼으로 설정해서 가져옴
df
```

```python
df = pd.read_csv('score.txt', sep='\t')
df.set_index('지원번호', inplace=True)
df
```

#### 엑셀 파일 열기 

```python
df = pd.read_excel('score.xlsx')
df
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```
