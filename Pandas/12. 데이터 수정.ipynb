### 12. 데이터 수정 

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

***
#### column 수정 

```python
df['소속'].replace({'떡잎유치원': '떡잎마을방범대', '미소속':'업음'})
```

```python
df['소속'].replace({'떡잎유치원': '떡잎마을방범대'}, inplace= True)
df
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['소속'] = df['소속'] + '소속' # 소속데이터 + '소속'
df
```

***
#### column 추가 

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'Fail' # 결과 컬럼 추가하고 데이터는 Fail로 초기화
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass' # 총합이 400보다 큰 데이터에 대해서 결과를 Pass로 업데이트
df
```

***
#### column 삭제

```python
df.drop(columns=['총합']) # 총합 컬럼을 삭제
```

```python
df.drop(columns=['국어', '영어', '수학']) # 국어, 영어, 수학 컬럼을 삭제
```

#### Row 삭제 

```python
df.drop(index='4번') # 4번 학생 데이터 행을 삭제
```

```python
filt = df['수학'] < 80 # 수학 점수가 80점 미만인 학생 필터링
df[filt]
```

```python
df[filt].index
```

```python
df.drop(index=df[filt].index)
```
