## 3. Index
데이터에 접근할 수 있는 주소 값 

```python
import pandas as pd

data = {
    '이름' : ['신짱구', '신형만', '봉미선', '신짱아', '훈이', '철수', '유리', '맹구'],
    '소속' : ['떡잎유치원', '떡잎상사', '없음', '없음', '떡잎유치원', '떡잎유치원', '떡잎유치원', '떡잎유치원'],
     '키' : [80, 184, 168, 60, 75, 80, 74, 85],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}

df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번'])
df
```

```python
df.index
```

### Index 이름 설정 

```python
df.index.name = '지원번호'
df
```

### Index 초기화 

```python
df.reset_index()
```

```python
df.reset_index(drop=True) # 원래 쓰던 '지원번호' 인덱스 삭제, 실제 데이터에는 반영되지 않음 !
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True) #실제 데이터에 바로 반영
```

```python
df
```

### Index 설정
지정한 column으로 index 설정 

```python
# df.set_index('이름')
df.set_index('이름', inplace= True)
```

```python
df
```

### Index  정렬
Index를 기준으로 오름차순, 내림차순 정렬 

```python
df.sort_index() # 인덱스로 오름차순 정렬
```

```python
df.sort_index(ascending=False) # 내림차순으로 정렬
```
