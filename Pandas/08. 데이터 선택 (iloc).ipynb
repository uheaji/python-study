### 8. 데이터 선택 (iloc)
위치를 이용하여 원하는 행에서 원하는 col 선택 

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.iloc[0] # 0번째 인덱스에 해당하는 데이터를 가져옴
```

```python
df.iloc[3] # 3번째 인덱스에 해당하는 데이터를 가져옴
```

```python
df.iloc[0:5] # 인덱스 0~4번째까지의 데이터를 가져옴
```

```python
df.iloc[0, 1] # 0번째 위치의 1번째(소속) 데이터를 가져옴
```

```python
df.iloc[0, 2] # 0번째 위치의 2번째(zl) 데이터를 가져옴
```

```python
df.iloc[[0, 1], 2] # 인덱스 0,1번째 위치 학생의 2번째(키) 데이터
```

```python
df.iloc[[0, 1], [3, 4]] # 인덱스 0,1번째 위치 학생의 3번째(국어), 4번째(영어) 데이터
```

```python
df.iloc[0:5, 3:6]
```
