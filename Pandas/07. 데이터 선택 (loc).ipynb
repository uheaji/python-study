### 7. 데이터 선택 (loc)
이름을 이용하여 원하는 행에서 원하는 col 선택 

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.loc['1번'] # 인덱스 1번에 해당하는 전체 데이터를 가져옴
```

```python
df.loc['5번'] # 인덱스 5번에 해당하는 전체 데이터를 가져옴
```

```python
df.loc['1번', '국어'] # 인덱스 1번에 해당하는 국어 데이터를 가져옴
```

```python
df.loc[['1번', '2번'], '영어'] # 인덱스 1번, 2번에 해당하는 영어 데이터를 가져옴
```

```python
df.loc[['1번', '3번'], ['영어', '수학']] # 인덱스 1번, 3번에 해당하는 영어, 수학 데이터를 가져옴
```

```python
df.loc['1번':'5번', '국어':'사회'] # 슬라이싱: 인덱스 1번부터 5번까지, 국어부터 사회까지의 데이터를 가져옴 (슬라이싱은 대괄호 필요없음)
```
