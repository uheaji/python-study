## DataFrame
2차원 데이터 (Series들의 모음)

### Data 준비
사전(dict) 자료구조를 통해 생성

예) 짱구는 못말려 주요 인물에 대한 데이터

```python
data = {
    '이름' : ['신짱구', '신형만', '봉미선', '신짱아', '훈이', '철수', '유리', '맹구'],
    '소속' : ['떡잎유치원', '떡잎상사', '없음', '없음', '떡잎유치원', '떡잎유치원', '떡잎유치원', '떡잎유치원'],
     '키' : [80, 184, 168, 60, 75, 80, 74, 85],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}

data
```

```python
type(data)
```

```python
data['이름']
```

### DataFrame 객체 생성 

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df
```

### 데이터 접근 

```python
df['이름']
```

```python
df['키']
```

```python
df[['이름', '키']]
```

### DataFrame 객체생성 (index 지정) 

```python
df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번'])
df
```

### DataFrame 객체생성 (column 지정)
data 중에서 원하는 column만 선택하거나, 순서변경 가능 

```python
df = pd.DataFrame(data, columns=['이름', '소속', '키'])
df
```
